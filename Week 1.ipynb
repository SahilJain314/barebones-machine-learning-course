# Week 1: Basic functions used in Machine Learning and an Overview

## Activation Functions

```python
import numpy as np  
import matplotlib.pyplot as plt
import math

def graph(function, x_range):  
    x = []
    for i in np.arange(x_range[0], x_range[1]+1, .25):
        x.append(i)
    
    if(function == 'sigmoid'):
        y = sigmoid(x)
    elif(function == 'tanh'):
        y = tanh(x)
    elif(function == 'relu'):
        y = relu(x)
    
    plt.plot(x, y)  
    plt.show()
```

#### The Sigmoid function

```python
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a

plt.axhline(0, color="gray",ls ="dashed", lw =1.5)
plt.axvline(0,color="gray",ls ="dashed", lw =1.5)
plt.axhline(1, color="gray",ls ="dashed", lw =1.5)
graph('sigmoid', [-5,5])
```

#### The Tanh function

```python
def tanh(x):
    a = []
    for item in x:
        a.append(math.tanh(item))
    return a

plt.axhline(0, color="gray",ls ="dashed", lw =1.5)
plt.axvline(0, color="gray",ls ="dashed", lw =1.5)
plt.axhline(1, color="gray",ls ="dashed", lw =1.5)
plt.axhline(-1, color="gray",ls ="dashed", lw =1.5)

graph('tanh', [-5,5])
```

#### The ReLU (Rectified Linear Unit) function

```python
def relu(x):
    a = []
    for item in x:
        if(item < 0):
            a.append(0)
        else:
            a.append(item)
    return a

plt.axhline(0, color="gray",ls ="dashed", lw =1.5)
plt.axvline(0, color="gray",ls ="dashed", lw =1.5)

graph('relu', [-5,5])
```
